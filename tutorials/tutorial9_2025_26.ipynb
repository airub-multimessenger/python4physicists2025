4. The code below has the aim to replace all negative numbers of ``a`` with zero. We want to replace the following code and use NumPy instead. How can we achieve this in the shortest way?

```python
a = [6,3,7,9,4,2,6,8,0,-1,6,3,7,0,4,3,-4]

for i in range(len(a)):
    if a[i] < 0:
        a[i] = 0

print(a)
```
