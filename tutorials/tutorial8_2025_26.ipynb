`````python
with open("students.csv", "w", newline="") as file:
    
    writer = csv.writer(file)

    writer.writerow(["Name", "Age", "Grade"]) # Header

    writer.writerow(["Alice", 20, "A"]) # Row 1
    writer.writerow(["Bob", 22, "B"]) # Row 2
`````

# Small tasks on ````numpy````

Numpy is a Python library that makes working with large amounts of numbers fast and easy. It is especially useful for scientific computing and data analysis. It allows mathematical operations to be performed quickly without writing loops and lets you work with many numbers at once, like vectors and matrices, in a simple and efficient way.

`````python
import numpy as np
`````

3. We want to calculate the scalar product of two vectors A and B. Try to rewrite the following code using numpy.

`````python
A = [0, -3, 5]
B = [-1, 2, 0]

c = 0
for i in range(len(A)):
    c += A[i] * B[i]
print(c)
`````
