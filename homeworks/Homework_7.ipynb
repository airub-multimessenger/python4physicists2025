## Mini Simulation: Compton Scattering

Compton scattering describes the phenomenon where a photon collides with an electron and loses energy, resulting in a longer wavelength.

The classical Compton formula is:

$$\Delta \lambda = \lambda_2 - \lambda_1 = \frac{h}{m_e c} (1 - cos(\Theta))$$

$\lambda_1$: initial wavelength of the photon

$\lambda_2$: scattered wavelength

$h = 6.626⋅10^{−34}$ J (Planck’s constant)

$m_e = 9.109⋅10^{−31}$ kg (electron rest mass)

$c = 3⋅10^8$ m/s (speed of light)

$\Theta$: scattering angle of the photon
